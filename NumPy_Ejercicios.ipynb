#### Implemente el algoritmo k-NN (*k Nearest Neighbours*)

*Indicaciones: Este algoritmo de aprendizaje supervisado consiste en asignar, a cada elemento, la media de las labels de los k vecinos más cercanos que tenga dicho elemento.*


*Veámoslo sobre un ejemplo: Supongamos que tenemos una serie de variables comportamentales de estudiantes (visualizaciones, ediciones, creaciones y descargas):*

```python
array([[19,  7, 12,  4],
       [19,  0,  7,  9],
       [17, 12, 17,  6],
       [ 8, 10,  1, 12],
       [ 9,  6,  0, 15],
       [11, 18,  1, 14],
       [16,  8, 12,  1],
       [ 2, 13,  6, 17],
       [ 0,  7, 13,  8],
       [15,  9, 10, 15]])
```

*Asociado a estos datos, tenemos además una calificación del estudiante a final de curso:*

```python
array([[6],
       [4],
       [7],
       [4],
       [3],
       [9],
       [9],
       [5],
       [3],
       [4]])
```

*A los datos que usamos para tratar de estimar las calificaciones los llamaremos features, y a la nota en sí la label del problema. Cada fila es el registro de un alumno, por lo que entendemos que la información de los estudiantes la podemos representar en un espacio 4-dimensional.*

*Bajo estas circunstancias, el algoritmo sigue estos pasos:*

1.  *Para un cierto estudiante sin calificación, calculamos la distancia (en el mencionado espacio 4-dimensional) del estudiante sin calificar a cada uno de los calificados.*

2.  *Habiendo prefijado el parámetro k, escogemos las notas de los k estudiantes más cercanos.*

3.  *Hacemos la media (moda en caso discreto) de estos datos, y esa será la estimación que tengamos del alumno.*


````python
#Creamos función que, dados dos arrays 1D, nos devuelva la distancia euclídea de los mismos
def euclidean_distance(vec1, vec2):
  #Algunas comprobaciones para mejorar la búsqueda de errores en código son:
  #if (type(vec1)!=np.ndarray) | (type(vec2)!=np.ndarray):
  # raise ValueError('El formato de alguno de los vectores de entrada no es np.array.')
  #if vec1.shape!=vec2.shape:
  # raise ValueError('Los arrays de entrada no tienen la misma forma.')
  return np.sqrt(np.sum((vec1-vec2)**2, axis = -1))
````

````python
#Fijamos número de "vecinos" en los que basar la nota
k = 2

#Obtenemos features
features = np.array([[19,  7, 12,  4],
                     [19,  0,  7,  9],
                     [17, 12, 17,  6],
                     [ 8, 10,  1, 12],
                     [ 9,  6,  0, 15],
                     [11, 18,  1, 14],
                     [16,  8, 12,  1],
                     [ 2, 13,  6, 17],
                     [ 0,  7, 13,  8],
                     [15,  9, 10, 15]])

#Obtenemos label
label = np.array([[6],[4],[7],[4],[3],[9],[9],[5],[3],[4]])

#Fijamos un alumno nuevo
new_student = np.array([15, 6, 5, 2])

#Calculamos distancias
distances = np.zeros(label.shape) #Inicializamos un array para contener las distancias
for i in np.arange(label.shape[0]):
  distances[i] = euclidean_distance(features[i,:], new_student)

#Obtenemos los índices de los k vecinos más próximos
optimal_idx = np.argpartition(distances.reshape(-1), k)[:k]

#Mostramos resultado
label_new_student = np.round(np.mean(label[optimal_idx]),2)
print('La predicción para el nuevo estudiante es: {}'.format(label_new_student))
````

Juego de Adivinanzas con NumPy  
Instrucciones: El programa ha generado un número aleatorio entre 1 y 100.   
Tienes 10 intentos para adivinar el número. 
Después de cada intento, recibirás una pista si tu adivinanza es mayor o menor que el número.   
Si adivinas el número, el programa te felicitará y terminará.   
Si no adivinas el número en 10 intentos, el programa te dirá cuál era el número.    
Para salir del juego, escribe 'exit' en lugar de un número.

````python
#Importamos libreria
import random 

#Generar un número aleatorio entre 1 y 100
numero_aleatorio=random.randint(1,100)           #lo nombro para poder usarlo y compararlo con lo que el usuario introduzca
print(numero_aleatorio)                          #uso randint para incluir el 1 y el 100 en las posibilidades #luego le quito el print
print(type(numero_aleatorio))                    #quiero ver el tipo de dato, para asegurarme que es el mismo tipo que ingrese el usuario

#Creamos un contador para los intentos de adivinar el número
intento=0

#Bucle de respuesta al usuario
for intento in range (1,11):
    #Solicitamos al usuario que escribe su número
    respuesta_usuario=input("Introduce un número entre 1 al 100. Si aciertas el número secreto, ¡Te llevas el bote!. Si quieres parar el juego, escribe la palabra exit")
   
    #Revisamos si la respuesta es "exit" para parar el juego
    if respuesta_usuario=="exit":                
        print("Es una pena que no lo pruebes de nuevo, ¡Nos vemos la próxima vez!")
        break
    
    #Si el usuario pone algo distinto a "exit", convertimos su entrada en un número entero para poder compararlo posteriormente
    numero_usuario=int(respuesta_usuario)         #como el número alatorio es entero, necesitamos que el del usuario lo sea también
    print(numero_usuario)
    print(type(numero_usuario))                  
        
    if numero_usuario==numero_aleatorio:
        print("¡Felicidades, has ganado el bote!")
        break
    
        
    if numero_usuario!=numero_aleatorio:
        print(f"Has fallado, pero no te preocupes, te quedan",(10-intento),"intentos.")   #restaremos a 10 (máximo de intentos), el número de intentos ya hechos
        
    
    if intento==10:
        print("Se han acabado tus intentos, suerte la próxima vez")    #Para el final del bucle, ponemos un mensaje distinto


````
